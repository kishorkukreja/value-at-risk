# What is Tail Risk?
I want to start by defining tail risk. 

In the my first story in this series I discussed how you can use the first four moments of a statistical distribution to find the volatility risk of an asset or a portfolio. This volatility risk (AKA standard deviation, σ) presents itself in the form of a distribution, where the deviation closest to the mode is most likely to happen, and the deviation farthest from the mode on either end of the distribution, called the tails, are least likely to happen. This means that large gains or losses are less likely to happen than small gains or losses.

Tail risk focuses on the probability and magnitude of the left tail - the condition of extreme negative returns. There are many different approaches to quantifying tail risk, but they can generally be categorized into three methodologies:

1. Historical
2. Parametric
3. Simulated

As the names suggests, historical approaches look at past data to find the magnitude and probability of extreme negative returns. The downside to this approach is that the future is not the same as the past, so a historical approach would not consider novel events (such as black swan financial crises) that could "outperform" historical risk estimates.

A parametric approach calculates tail risk as a function of multiple factors using a parametric model. Later we will implement a percent point function that considers the mean return, standard deviation, and confidence level. A parametric model's power is based on the data used to create the model. If the model is only fit to a limited range of historical data, it could have the same weakness as a historical approach.

The last category is the simulated approach, which generates numerous potential outcomes based on random simulations of data to estimate what the risk would be in each potential outcome. This approach is limited by the complexity of the simulations. Actual movements in price action are not truly random (otherwise technical patterns would be of no use), so a series of truly random simulations might not be as accurate for predicting risk as a simulation that accounts for typical factors in price movements like earnings reports or interest rate changes.

Now that you have an understanding of what tail risk is and the different approaches to estimating it, let's look at some specific methods and apply them to a hypothetical portfolio.

# The Drawdown of Your Portfolio
One of the most common criteria used to evaluate a portfolio's performance is the maximum drawdown. The maximum drawdown is your greatest percentage loss from your highest cumulative return. It's a clear indication of the effectiveness of your risk management strategy.

Before I go over the calculation of the maximum drawdown, let's first import the usual libraries and create a portfolio using Empyrial. I've used this package several times now for portfolio analysis, but if you haven't used it before, check out my story "What's the Best Portfolio Optimization Method for You?" I go over all of the details of what's available in this module in that story.

We'll start with a diversified portfolio using the mean-variance optimizer and back-test on the last five years.

```python
import pandas as pd
import matplotlib.pyplot as plt
from empyrial import empyrial, Engine
import numpy as np
from scipy.stats import norm
import math

portfolio_MEANVAR = Engine(
    start_date = "2016-10-01",
    portfolio = ['GLD', 'SPY', 'BND', 'TQQQ', 'GBTC', 'SHOP', 'BAC', 'BA'],
    benchmark = ['TQQQ'],
    optimizer = "MEANVAR",
    #optimizer = 'HRP',
    rebalance = "1y",
    #risk_manager = {"Max Drawdown" : -0.05}
)

empyrial(portfolio_MEANVAR)
print(f'The maximum drawdown of the portfolio is {empyrial.MD}.')
```

Empyrial's back-testing function generates a report that shows the maximum drawdown, the conditional daily value at risk, an underwater plot showing the drawdown, and a plot showing the top five drawdown periods. If you'd like to take a look at what's going on underneath the hood of Empyrial, check out the GitHub repo here. We will be checking our calculations against the values provided in this report.

You can pull the cumulative returns of your portfolio from empyrial.creturns. Then use the .maximum() and .accumulate() methods on these returns to grab the highest cumulative return. The drawdown can be calculated as one minus the ratio of cumulative returns to the running_max.

```python
creturns = empyrial.creturns
running_max = np.maximum.accumulate(creturns) # record the maximum cumulative return
running_max[running_max<1] = 1 # put a limit on max cumulative return at 1
drawdown = ((creturns)/running_max - 1)*100 # calculate the drawdown
print(f'The maxmimum drawdown is {round(min(drawdown),2)}%')
drawdown.plot()
plt.title('Five Year Drawdowns')
plt.xlabel('Year')
plt.ylabel('% Drawdown')
plt.show()
```

This returns a maximum drawdown of -27.43%, which is identical to the drawdown reported by Empyrial. Our underwater plot also matches the underwater plot generated by Empyrial.

A drawdown of -27% is extreme. Try playing around with the risk manager and optimizer parameters to lower the maximum drawdown. I found that with no risk management, TQQQ vastly outperforms the portfolio, but the two actually perform similarly with risk management.

# Historical Value at Risk

The next method of analyzing tail risk is to calculate the historical value at risk. This metric is the minimum loss you can expect at a given quantile (also known as the confidence level).

The historical value at risk can easily be found with the np.percentile() method. Pass in the series of daily returns and the quantile as arguments.

Below I show an example of finding the historical value at risk at the 95% confidence level. To do this, I pass in 5% as the quantile argument. The result, -1.58%, means that in 5% of days in the past 5 years my losses exceeded -1.58%. The corollary to this finding is that I can be 95% certain that my losses will not exceed -1.58%. This value is represented by the vertical red line below.

```python
returns = empyrial.returns * 100
var_95 = np.percentile(returns, 5) # find the historical value at risk 
print(f' The 95% historical daily value at risk is {round(var_95,2)}%')
sorted_rets = sorted(returns)
plt.hist(sorted_rets, bins = 40)
plt.axvline(x=var_95, color='r', linestyle='-', label="VaR 95: {0:.2f}%".format(var_95))
plt.title('Distribution of 5 Year Daily Portfolio Returns')
plt.xlabel('% Daily Return')
plt.ylabel('Number of Days per % Return')
plt.show()
```

I want to note that historical VaR is frequently estimated at 95%, 99%, and 99.9% confidence levels. The expected losses will increase significantly as you increase the confidence level. I chose the 95% confidence level because that is the default for the conditional value at risk in Empyrial. However, conditional value at risk is slightly different than historical value at risk, so let's look into that.

# Conditional Value at Risk

Think of historical value at risk as the lower limit of your expected losses for your chosen confidence level. Conditional value at risk, also known as expected shortfall, is the average expected loss at a given confidence level. Therefore, calculating the CVaR is as simple as finding the mean of losses exceeding the historical VaR. The CVaR is represented by the blue vertical line in the histogram below.

```python
cvar_95 = np.mean(returns[returns<var_95])
print(f'The 95% conditional value at risk is {round(cvar_95,2)}%.')

plt.hist(sorted_rets, bins = 40)
plt.axvline(x=var_95, color='r', linestyle='-', label="VaR 95: {0:.2f}%".format(var_95))
plt.axvline(x=cvar_95, color='b', linestyle='-', label='CVaR 95: {0:.2f}%'.format(cvar_95))
plt.title('Distribution of 5 Year Daily Portfolio Returns')
plt.xlabel('% Daily Return')
plt.ylabel('Number of Days per % Return')
plt.show()
```

The 95% expected shortfall based on the past five years of returns is -2.55%, so 5% of the time we can expect that as the average loss.

You may have noticed that this calculation for the expected shortfall does not match the -2.0% daily value at risk produced by the Empyrial report. Under the hood, Empyrial uses the .value_at_risk() function from the Quantstats package, passing in a default confidence level of 95%. The Quantstats function then calculates the parametric conditional value at risk using the percent point function from SciPy - norm.ppf().

Note: I didn't know what a percent point function is, so I googled it. The NIST defines is as:

"For the percent point function, we start with the probability and compute the corresponding x for the cumulative distribution."

The percent point function can also be said to return the z-score for a given probability, where the z-score is "a numerical measurement that describes a value's relationship to the mean of a group of values."

```python
mu = np.mean(returns)
vol = np.std(returns)
confidence_level = 0.95
var_95 = norm.ppf(1-confidence_level, mu, vol)
print(f'Mean: {str(round(mu,2))}%\nVolatility: {str(round(vol,2))}%\nVaR(95): {str(round(var_95,2))}%')
```

Now we have a CVaR of -1.68%. This still does not match the output of the Empyrial report, and honestly, I'm not sure why. You can investigate the source code for Empyrial here, and the source code for Quantstats here. If you figure out why my numbers are off, please let me know. But otherwise, you might feel more confident about your calculation for expected shortfall rather than the output of the Empyrial report.

# Scaling Value at Risk With Time

Our previous calculations of the value at risk express an extreme expectation for the portfolio's potential loss on any given day. However, markets often move in trends that could span several days, weeks, or even months. For this reason, you may find it prudent to also estimate your value at risk for an extended period of time.

Scaling value at risk is done in the same way as scaling volatility calculations - multiply by the square root of time.

```python
# Aggregate forecasted VaR
forecasted_values = np.empty([100, 2])

# Loop through each forecast period
for i in range(100):
    # Save the time horizon i
    forecasted_values[i, 0] = i
    # Save the forecasted VaR 95
    forecasted_values[i, 1] = var_95 * np.sqrt(i + 1)
    
# Plot the results
print(f'The forecasted value at risk after one-hundred consecutive of losses exceeding the 5th percentile is {round(forecasted_values[-1,-1],2)}%')
plt.plot(forecasted_values[:,0], abs(forecasted_values[:,1]))
plt.title('100 Day Scaled Value At Risk')
plt.xlabel('Number of Days')
plt.ylabel('% Value At Risk')
plt.show()
```

And with that we're able to see how our value at risk scales across one-hundred days. Based on this calculation, the portfolio would lose 16.8% if it suffered losses exceeding the 5th percentile of returns every day for one-hundred days.

# Random Walk Simulations

I just learned that stochastic is just a fancy term for random, so don't be afraid if someone asks you about stochastic simulation because it's one and the same as what we're about to go over. A random or stochastic walk is a way of generating random results based on a distribution with the same mean and standard deviation as a sample group. 

In the example below, we set the length (T) of the ndarray that will be generated to be the same length as the returns we received from Empyrial. We are using empyrial.returns because the earlier returns variable contains the returns as a percentage - we need the returns in raw decimal format. S0 is the starting value of the portfolio.

After calculating the mean and standard deviation of our returns, we create the ndarray of random returns, add one, and store this in the rand_rets variable. You can then calculate the forecasted values of the portfolio based on the cumulative product of the random returns times the initial value.

Notes: 

1. We add one to the random returns to convert the returns into multiplicative factors for when we calculate the cumulative product. If we did not add one, the repeated multiplication of two fractions less than one will create increasingly smaller decimals. Eventually the cumulative product of returns would converge to zero without the addition of one.

2. If you are working with annualized average returns and annualized volatility, you can use the commented line with the daily_returns variable to generate random daily returns such that T is set to the number of trading days within the sample timeframe.

3. We are using a normal distribution of randomly generated returns. In my previous stories, I mentioned that financial returns are typically non-normal. Therefore, our random returns are not representative of an actual distribution of returns; instead they serve as a baseline for analyzing random returns. Keep this in mind when interpreting the results of the cumulative returns.

```python
# Set the simulation parameters
T = len(empyrial.returns)
S0 = 1000
mu = np.mean(empyrial.returns)
vol = np.std(empyrial.returns)

# Add one to the random returns
rand_rets = np.random.normal(mu, vol, T) + 1
# Forecasted random walk
forecasted_values = rand_rets.cumprod() * S0

# daily_returns=np.random.normal((mu/T),vol/math.sqrt(T),T)+1

# Plot the random walk
plt.plot(range(0, T), forecasted_values)
plt.title('5 Year Simulated Random Returns')
plt.xlabel('Number of Days')
plt.ylabel('Portfolio Value')
plt.show()
plt.hist(empyrial.returns,80)
plt.hist(rand_rets-1,80)
plt.title('5 Year Random Returns Distribution')
plt.xlabel('% Returns')
plt.ylabel('Number of Days per % Return')
plt.show()

var_95 = np.percentile(rand_rets-1, 5) # find the historical value at risk 
print(f'The simulated historical value at risk is{round(var_95*100,2)}%.')
cvar_95 = np.mean(rand_rets[rand_rets-1<var_95]-1)
print(f'The simulated conditional value at risk is{round(cvar_95*100,2)}%.')
mu = np.mean(rand_rets-1)
vol = np.std(rand_rets-1)
confidence_level = 0.95
cvar_95 = norm.ppf(1-confidence_level, mu, vol)
print(f'Mean: {str(round(mu*100,2))}%\nVolatility: {str(round(vol*100,2))}%\nCVaR(95): {str(round(cvar_95*100,2))}%')
```

From the results we can see that the historical VaR, CVaR, and parametric CVaR based on the random distribution closely match that of the original sample.

# Monte Carlo Simulations

The final approach that I want to demonstrate is to simulate returns data using a Monte Carlo simulation. This type of simulation works by creating multiple random walk simulations (anywhere from 10 to 10,000 or more). The results exemplify a broad range of possible outcomes and ultimately serve the purpose of conduct a more robust statistical analysis.

```python
# loop the the number of iterations you want in your Monte Carlo Simulation
# More iterations will require more time to compute
sim_returns = pd.DataFrame()

for i in range(10):
    # Add one to the random returns
    rand_rets = np.random.normal(mu, vol, T) + 1
    sim_returns[i] = rand_rets-1
    # Forecasted random walk
    forecasted_values = rand_rets.cumprod() * S0
    # Plot the random walk
    plt.plot(range(0, T), forecasted_values)
plt.title('5 Year Simulated Random Returns')
plt.xlabel('Number of Days')
plt.ylabel('Portfolio Value')
plt.show()
plt.hist(sim_returns*100, bins = 20)
plt.title('5 Year Random Returns Distribution')
plt.xlabel('% Returns')
plt.ylabel('Number of Days per % Return')
plt.show()
# Calculate the VaR(95)
var_95 = np.percentile(sim_returns, 5)
print("Simulated Historical VaR(95): ", round(100*var_95, 2),"%")
cvar_95 = np.mean(np.mean(sim_returns[sim_returns<var_95]))
print("Simulated Conditional VaR(95): ", round(100*cvar_95, 2),"%")
```
